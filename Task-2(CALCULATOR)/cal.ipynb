```python
import tkinter.font
from tkinter import *
import math

expression = ''
value = ''
ans = ''

def clear_all():
    global expression, value, ans
    expression = ''
    value = ''
    ans = ''
    equation.set(expression)

def sign(num):
    return 1 if num > 0 else -1 if num < 0 else 0

def clear_back():
    result1 = ""
    result2 = ""
    global equation, expression, value, ans

    expression = input_field.get()
    temp_expression = list(expression)
    temp_value = list(value)

    if not value:
        temp_expression = []
        temp_value = []

    special_functions = ["asin(", "acos(", "atan(", "log("]
    for function in special_functions:
        if expression.endswith(function):
            temp_expression = temp_expression[:-len(function)]
            temp_value = temp_value[:-len(function)]

    functions_to_length = {
        'sin(': 4, 'cos(': 4, 'tan(': 4, 'sgn(': 4, 'ln(': 3, 'e^': 2, '^': 1, '√': 1, 'π': 1, 'e': 1, '%': 1
    }

    for function, length in functions_to_length.items():
        if expression.endswith(function):
            temp_expression = temp_expression[:-length]
            temp_value = temp_value[:-length]

    if temp_expression:
        temp_expression.pop()
    if temp_value:
        temp_value.pop()

    expression = ''.join(temp_expression)
    equation.set(expression)

    value = ''.join(temp_value)
    try:
        ans = str(eval(value))
    except:
        ans = "Invalid Expression"

def press_button(num):
    global expression, value, ans
    expression += str(num)
    equation.set(expression)

    if num in ["1", "2", "3", "4", "5", "6", "7", "8", "9", "0", "(", ")", "00"]:
        value += num
        try:
            ans = str(eval(value))
        except:
            ans = "Invalid Expression"
    elif num in ["+", '-', '/', '*', '.', '1/', 'sgn(']:
        value += num
    elif num in ['asin(', 'acos(', 'atan(', 'sin(', 'cos(', 'tan(']:
        value += 'math.' + num
    elif num == '^':
        value += '**'
    elif num == '%':
        value += '/100'
        try:
            ans = str(eval(value))
        except:
            ans = "Invalid Expression"
    elif num == '^2':
        value += '**2'
        try:
            ans = str(eval(value))
        except:
            ans = "Invalid Expression"
    elif num == '^3':
        value += '**3'
        try:
            ans = str(eval(value))
        except:
            ans = "Invalid Expression"
    elif num == '√(':
        value += 'math.sqrt('
    elif num == 'e':
        value += 'math.e'
        try:
            ans = str(eval(value))
        except:
            ans = "Invalid Expression"
    elif num == 'π':
        value += 'math.pi'
        try:
            ans = str(eval(value))
        except:
            ans = "Invalid Expression"
    elif num == 'log(':
        value += 'math.log10('
    elif num == 'ln(':
        value += 'math.log('
    elif num == 'e^':
        value += 'math.e**'

def compute_result():
    global ans, value, expression

    if not value:
        ans = ""
    equation.set(ans)
    ans = ""
    value = ""
    expression = ""

root = Tk()
root.title("Scientific Calculator")
root.resizable(False, False)

calculator_frame = Frame(root)
calculator_frame.grid()
calculator_frame.configure(bg="#FF5733")  # Change background color to light red

equation = StringVar()
input_field = Entry(
    calculator_frame,
    textvariable=equation,
    width=60,
    font=("Comic Sans MS", 15),
    bd=10,
    justify=LEFT,
    state=DISABLED,
    disabledbackground="white",
    disabledforeground="black"
)
input_field.insert(0, "0")
input_field.grid(row=0, columnspan=8)

font = tkinter.font.Font(size=12, weight="bold", family='Helvetica')
h = 2
w = 7
actvbgnd = 'white'
bg1 = '#FF5733'  # Light Red Hex Color
bg2 = "#00008B"  # Dark Blue Hex Color
bg3 = "burlywood2"
bg4 = "tan1"
fg1 = "white"
fg2 = "black"

numberpad = [7, 8, 9, 4, 5, 6, 1, 2, 3]
i = 0
for j in range(3):
    for k in range(3):
        Button(calculator_frame, command=lambda x=str(numberpad[i]): press_button(x), text=str(numberpad[i]),
               bg=bg1,  # Change button background color to light red
               fg=fg2, activebackground=actvbgnd,
               height=h, width=w, font=font).grid(row=j + 2, column=k)
        i += 1

r = 5
c = 7
Button(calculator_frame, command=lambda: press_button('0'), text="0", bg=bg1,  # Change button background color to light red
       fg=fg2, activebackground=actvbgnd,
       height=h, width=w, font=font).grid(row=r, column=c - 7)
Button(calculator_frame, command=lambda: press_button('00'), text="00", bg=bg1,  # Change button background color to light red
       fg=fg2, activebackground=actvbgnd,
       height=h, width=w, font=font).grid(row=r, column=c - 6)
Button(calculator_frame, command=clear_back, text="C", bg=bg2,  # Change button background color to dark blue
       fg=fg2, activebackground=actvbgnd,
       height=h, width=w, font=font).grid(row=r - 4, column=c - 7)
Button(calculator_frame, command=clear_all, text="AC", bg=bg2,  # Change button background color to dark blue
       fg=fg2, activebackground=actvbgnd,
       height=h, width=w, font=font).grid(row=r - 4, column=c - 6)
Button(calculator_frame, command=lambda: press_button('.'), text="•", bg=bg3,
       fg=fg2, activebackground=actvbgnd,
       height=h, width=w, font=font).grid(row=r, column=c - 5)
Button(calculator_frame, command=lambda: press_button('+'), text="+", bg=bg3,
       fg=fg2, activebackground=actvbgnd,
       height=h, width=w, font=font).grid(row=r - 2, column=c - 4)
Button(calculator_frame, command=lambda: press_button('-'), text="–", bg=bg3,
       fg=fg2, activebackground=actvbgnd,
       height=h, width=w, font=font).grid(row=r - 3, column=c - 4)
Button(calculator_frame, command=lambda: press_button('/'), text="/", bg=bg3,
       fg=fg2, activebackground=actvbgnd,
       height=h, width=w, font=font).grid(row=r - 4, column=c - 5)
Button(calculator_frame, command=lambda: press_button('*'), text="✶", bg=bg3,
       fg=fg2, activebackground=actvbgnd,
       height=h, width=w, font=font).grid(row=r - 4, column=c - 4)
Button(calculator_frame, command=compute_result, text="=", bg=bg2,  # Change button background color to dark blue
       fg=fg2, activebackground=actvbgnd,
       height=2 * h, width=w, font=font, pady=10).grid(row=r - 1, column=c - 4, rowspan=2)

functions = ['(', ')', '%', 'asin', 'sin', 'log', 'x^2', 'acos', 'cos', 'ln', 'x^3', 'atan', 'tan', 'e^x', '1/x', 'x^y', 'e', "π", '√x', 'sgn']
functions_values = ['(', ')', '%', 'asin(', 'sin(', 'log(', '^2', 'acos(', 'cos(', 'ln(', '^3', 'atan(', 'tan(', 'e^', '1/', '^', 'e', "π", '√(', 'sgn(']
i = 0
for j in range(5):
    for k in range(4):
        Button(calculator_frame, command=lambda x=functions_values[i]: press_button(x), text=functions[i],
               bg=bg4,  # Change button background color to dark blue
               fg=fg2, activebackground=actvbgnd,
               height=h, width=w, font=font).grid(row=j + 1, column=k + 4)
        i += 1

cal_row_size = 50
calculator_frame.rowconfigure(0, minsize=cal_row_size)
for i in range(1, 6):
    calculator_frame.rowconfigure(i, minsize=cal_row_size)

cal_column_size = 90
for i in range(8):
    calculator_frame.columnconfigure(i, minsize=cal_column_size)

root.mainloop()
```
